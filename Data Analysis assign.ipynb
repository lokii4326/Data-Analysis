```python
# ‚úÖ Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# ‚úÖ Step 2: Upload CSV file from your computer
print("üìÅ Please upload your Organization.csv file...")
uploaded = files.upload()

# ‚úÖ Step 3: Read the CSV file into a DataFrame
filename = next(iter(uploaded))  # gets the uploaded filename
data = pd.read_csv(filename)
print("‚úÖ File Loaded Successfully!")

# ‚úÖ Step 4: Preview data
print("\nüìÑ First 5 Rows:")
print(data.head())

print("\n‚ÑπÔ∏è Dataset Info:")
print(data.info())

# ‚úÖ Step 5: Check missing values and duplicates
print("\nüßº Missing Values:\n", data.isnull().sum())
print("üîÅ Duplicates Found:", data.duplicated().sum())

# ‚úÖ Step 6: Convert Gender to numeric if it exists
if 'Gender' in data.columns:
    data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
    print("\n‚úÖ Gender column converted to numeric.")

# ‚úÖ Step 7: Detect numeric columns
numeric_cols = data.select_dtypes(include='number').columns.tolist()
print(f"\nüìä Numeric columns: {numeric_cols}")

# ‚úÖ Step 8: Boxplots for numeric columns
for col in numeric_cols:
    plt.figure()
    sns.boxplot(y=data[col])
    plt.title(f'Boxplot - {col}')
    plt.show()

# ‚úÖ Step 9: Correlation heatmap
if len(numeric_cols) > 1:
    plt.figure(figsize=(8, 6))
    sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm')
    plt.title('üìà Correlation Heatmap')
    plt.show()
else:
    print("\n‚ÑπÔ∏è Not enough numeric columns to show correlation heatmap.")
```
